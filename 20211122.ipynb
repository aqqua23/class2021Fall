```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
amp = 1; sr = 10000; phase = 0

B4 = 493.88
D4 = 293.66
E4 = 329.63
F4 = 369.99
G4 = 392.99
A3 = 220.00
B3 = 246.94
G3 = 196.00
```

```python
# Love Poem

notes = [D4,F4,G4,0,G4,G4,G4,E4,E4,D4,G4,E4,0,\
         E4,D4,D4,0,F4,G4,0,G4,G4,G4,E4,E4,D4,G4,E4,0,\
         E4,D4,B4,G4,G4,0,E4,D4,B4,G4,G4,0,\
         G4,G4,G4,E4,E4,0,D4,G4,E4,D4,G3,A3,0,B3,G3,G3]

durs = [0.8,0.4,0.8,0.2,0.2,0.2,0.4,0.2,0.6,0.2,0.2,0.6,0.2, \
        0.2,0.2,0.8,0.2,0.4,0.6,0.2,0.2,0.2,0.4,0.2,0.6,0.2,0.2,0.6,0.2, \
        0.2,0.2,0.8,0.2,0.6,0.2,0.2,0.2,0.8,0.2,0.6,0.2, \
        0.2,0.2,0.6,0.2,0.6,0.2,0.2,0.2,0.6,0.2,0.2,1,0,0.4,0.2,0.2,0.6]
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th = phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all,s))
```

```python
ipd.Audio(s_all, rate=sr)
```
